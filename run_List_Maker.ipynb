```python
# {event}_{basin}_JPM_{sim}
basinSims = {
    '5P' :[1,100],
    '25P':[101,200],
    '50P':[201,300],
    '75P':[301,400],
    '95P':[401,500]
}

precip_eventList = [item for item in range(1, 645+1)]
eventList = []
for i in precip_eventList:
    pad = '0'
    n=4
    x = str(i).rjust(n, pad)
    eventList.append(x)

# myProject = Project.open(r"Z:\Amite\HEC_HMS_RE-Calibration\HMS_Model_AORC_TS_Matthew\Amite_Final_HMS_Model.hms")
```

```python
basinSims['5P'][1]
```

```python
runList = []
for event in eventList:
    for basin in basinSims.keys():
        for sim in range(basinSims[basin][0], basinSims[basin][1]+1):
            runList.append([event, basin, sim])
```

```python
len(runList)
```

```python
import pandas as pd
df = pd.DataFrame(runList, columns =['Event', 'Basin', 'Sim'])
df.to_csv('runList.csv')
```

```python
import glob
import pandas as pd
# rainfall_preprocessor file name list clean

in_dir = r"V:\projects\p00542_cpra_2020_lwi_t10\00_collection\data\rainfall\download"
f_in = glob.glob(fr"{in_dir}\**\*.nc", recursive=True)
out_dir = r"V:\projects\p00542_cpra_2020_lwi_t10\00_collection\data\rainfall\time_fixed_nc"
f_out = glob.glob(fr"{out_dir}\**\*.nc", recursive=True)

```

```python
import os 
f_in_tails = []
f_out_tails = []

for f in f_in:
    head, tail = os.path.split(f)
    f_in_tails.append(tail)

for f in f_out:
    head, tail = os.path.split(f)
    f_out_tails.append(tail)
```

```python
df_in = pd.DataFrame(f_in_tails, columns=['0'])
df_out = pd.DataFrame(f_out_tails, columns=['0'])
```

```python
df_all = df_in.merge(df_out.drop_duplicates(), on=['0'], 
                   how='left', indicator=True)
```

```python
df_clean = df_all[df_all['_merge'] == 'left_only']
```
